```python
def get_partition_pairs(ps: PS) -> list[(PS, PS)]:
    result = [(PS.empty(problem.search_space), PS.empty(problem.search_space))]
    
    for index in ps.get_fixed_variable_positions():
        value = ps.values[index]
        new_result = []
        new_result.extend([(left.with_fixed_value(index, value), right) for left, right in result])
        new_result.extend([(left, right.with_fixed_value(index, value)) for left, right in result])
        result = new_result
        
    # only keep the ones with the right size
    result = result[(len(result)//2):]
    
    result = result[:-1]
        
    # def pair_is_ordered(pair):
    #     return (pair[0].fixed_count() <= pair[1].fixed_count())
    
    #result = [pair for pair in result if pair_is_ordered(pair)]
    return result


ps= pss_to_evaluate[-1]
pairs = get_partition_pairs(ps)
for pair in pairs:
    print(pair)
    
print(len(pairs))
n = ps.fixed_count()
expected = 2**(n-1) - 1
print(f"{expected = }")
        
```
